```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
bootcamp_data = pd.read_csv(file, low_memory = False)
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
main_df = bootcamp_data.iloc[:,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
```

```python
# Change reading 0 and 1 to No and Yes, respectively
main_df.replace(to_replace = [0.0, 1.0], value = ["No", "Yes"], inplace = True)
```

```python
# Extract rows for only those who attended a bootcamp
attendee_df = main_df.loc[main_df["AttendedBootcamp"] == "Yes", :]
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
attendee_count = pd.DataFrame(attendee_df["BootcampName"].value_counts())
attendee_count.reset_index(inplace = True)
attendee_count.columns = ["BootcampName", "Attendee Count"]
attendee_count
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
rec_count = attendee_df.loc[attendee_df["BootcampRecommend"] == "Yes", :]
rec_count = pd.DataFrame(rec_count["BootcampName"].value_counts())
rec_count.reset_index(inplace = True)
rec_count.columns = ["BootcampName", "Rec Count"]
rec_count
```

```python
# Merge the two created data frames on the name of the bootcamp
full_df = pd.merge(attendee_count, rec_count, on = "BootcampName")
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
pct_rec = (full_df["Rec Count"] / full_df["Attendee Count"]) * 100
pd.options.display.float_format = '{:.2f}%'.format
full_df = full_df.sort_values(by = "Percent Rec", ascending = False)
full_df
```

```python
# Export to excel and remove index
full_df.to_csv("output/Project_Part_2_Output.csv", index = False)
```
